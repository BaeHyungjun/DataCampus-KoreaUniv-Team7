## 품사태깅 & 어미 제외

```python
han = Hannanum()
```

```python
# 앞에서부터 10만 개 데이터로만 한정
# naver_cleaned_copy = naver_cleaned.iloc[:100000, :].copy()
daum_cleaned_copy = daum_cleaned.iloc[:100000, :].copy()
```

```python
def stemming_body(row):
    '''기사 본문 품사태깅 후 어미가 아닌 형태소들만 추출하는 함수'''
    stem = []
    tagged = han.pos(row['body'])
    for _ in tagged:
        if _[1] not in ['E', 'EC', 'EF', 'EP', 'ET', 'S']:
            stem.append(_[0])
        
    if row['id'] % 10000 == 0:
        print('{0} id {1} is processed'.format(datetime.datetime.now(), row['id']))
        
    return stem
```

```python
def stemming_summary(row):
    '''기사 요약문 품사태깅 후 어미가 아닌 형태소들만 추출하는 함수'''
    stem = []
    tagged = han.pos(row['summary'])
    for _ in tagged:
        if _[1] not in ['E', 'EC', 'EF', 'EP', 'ET', 'S']:
            stem.append(_[0])
        
    if row['id'] % 10000 == 0:
        print('{0} id {1} is processed'.format(datetime.datetime.now(), row['id']))
        
    return stem
```

```python
stopwords2 = pd.read_csv('../../stopwords/stopwords_concat.csv')
stopwords_list2 = list(stopwords2['stopwords'].values)
len(stopwords_list2)
```

```python
stopwords_list = stopwords_list1 + stopwords_list2
print(len(stopwords_list))
stopwords_list = list(set(stopwords_list))
print(len(stopwords_list))
```

```python
def remove_stopwords(morphs_list):
    '''형태소(body_stem / summary_stem) 데이터를 받아 불용어를 처리하는 함수'''
    new_list = [x for x in morphs_list if x not in stopwords_list]
    return new_list
```

```python
news_stemmed.body_stem = news_stemmed.body_stem.apply(remove_stopwords)
```

```python
news_stemmed.summary_stem = news_stemmed.summary_stem.apply(remove_stopwords)
```

```python
least_list = least_body + least_summary
```

```python
print(len(least_list))
least_list = list(set(least_list))
print(len(least_list))
```

```python
def remove_least_morphs(morphs_list):
    '''형태소(body_stem / summary_stem) 데이터를 받아 가장 적게 나온 형태소들을 처리하는 함수'''
    new_list = [x for x in morphs_list if x not in least_list]
    return new_list
```

```python
news_stemmed_copy.body_stem = news_stemmed_copy.body_stem.apply(remove_least_morphs)
```

```python
news_stemmed_copy.summary_stem = news_stemmed_copy.summary_stem.apply(remove_least_morphs)
```
